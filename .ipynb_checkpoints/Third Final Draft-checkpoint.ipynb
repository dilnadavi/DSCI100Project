The usability of a dataset is largely important in dataset analysis. It is vital that our dataset is tidy in order to avoid one-time approaches that results in difficulty for reiteration and reuse -- a tidy dataset allows us to use the tools that the tidyverse library gives us with relative ease and familiarity. We see that this dataset generally meets the requirements to be tidy as each column is a variable, each row is an observation and each cell forms a single measurement. However, looking at the last observation, we see that the value for the age of participant 196 is "NA".

```r
players_NA <- players_data |> filter(is.na(Age))
players_NA
```

The decision to remove rows with missing measurments is a complicated process, as information that measurements are missing can itself be a useful observation. However, as shown above, there are only two rows with missing measurements for age, and thus, its removal will make further data analysis more manageable, as summary statistics are often highly affected by missing observations, while also not having an immense impact on the integrity of our data. Below, we wrangle the data to remove these observations. Here, we will also take a step ahead and transform subscribe as a factor for classification, which is to be explained later after our exploratory analysis.

```r
## wrangle the data
wrangled_players_data <- players_data |> drop_na(Age) |> mutate(subscribe = as.factor(subscribe))
tail(wrangled_players_data)
```

As can be seen, observations with missing measurements for age have been removed, resulting in two less rows. Now that this dataset is easier to perform analysis on, we can explore the relationships between our variables of interest. When considering the subscription status of a player, common sense tells us that those who subscribe to a gaming newsletter will play more hours than players who are not subscribed with reasons of dedications. To see if this assumption holds, we can calculate summary statistics about the average hours played for subscribed versus unsubscribed players. To have a stronger grasp about the relationship between subscription status and the total number of hours played, we can create a visualization to present patterns present in the data. As we are observing a single quantative variable for each group, a histogram is most appropriate. With the visualization, we can use the facet_grid() function to compare the count of subscribed versus unsubscribed players that falls into each bin.

```r
average_played_hours_per_subscribe <- wrangled_players_data |> group_by(subscribe) |> summarize(average_played_hours = mean(played_hours))
average_played_hours_per_subscribe
ggplot(wrangled_players_data, aes(x = played_hours)) +
  geom_histogram(binwidth = 30, fill = "skyblue", color = "black") +
  facet_grid(cols = vars(subscribe)) +
  labs(
    title = "Figure 1: Distribution of Hours Played by Subscription Status",
    x = "Total Hours Played (hours)",
    y = "Count of Players"
  )
```

Here, we see that there is much more data pertaining to subscribed users as opposed to unsubscribed users, indicating a possible over-representation of a group in the original dataset. However, we can also observe that the average total hours played for subscribed users is substantially greater by a factor of ~16 compared to unsubscribed players. Furthermore, we can also note there are no observations in the bins that exceed ~35 hours for unsubscribed players, while there are observations in the 100-200 total played hours for subscribed players -- this indicates that the range of total hours played is far greater for subscribed players, and as a premature assumption, one can observe that subscribed players seem to play more hours than unsubscribed players. This indicates that the number of hours played could potentially be a predictor for the subscription status of the player, and thus can meaningfully contribute to our prediction model.

Though we can theoretically start creating our prediction model with the single predictor of hours played, oftentimes, it is rare that a single predictor can successfully classify data without bias. Thus, finding another predictor for the subscription status can potentially create a stronger model. Looking at our dataset, one can ponder whether a player's age has an effect on their decision to subscribe to a newsletter. To analyze any such patterns, we can again summarize the mean and visualize these two variables.

```r
average_age_per_subscribe <- wrangled_players_data |> group_by(subscribe) |> summarize(average_age = mean(Age))
average_age_per_subscribe
ggplot(wrangled_players_data, aes(x = Age)) + 
  geom_histogram(binwidth = 30) + 
  facet_grid(cols = vars(subscribe)) +
  labs(
    title = "Figure 2:Age Distribution by Subscription Status",
    x = "Player Age",
    y = "Frequency")
```

With the summary statistic, we see that the average age of unsubscribed players is ~22, whereas the average age for subscribed players is ~19. Through these visualizations, we see that even with the disproportionate representation of subscribed players, there is considerably more data in the older-age bins for unsubscribed players than subscribed players. Moreover, we see that there is a substantial count of players falling into the younger-age bins for subscribed players as well, whereas there is no such observations for unsubscribed players. Thus, this analysis suggests that younger players are more likely to subscribe to the gaming newsletter, and thus, is a potential predictor for the player's subscription status.

With this, we have two predictors for our prediction model. As aforementioned, our report aims to create a model that can predict a player's subscription status. To do this, we need a classification algorithm. In this report, the algorithm of interest is the KNN Classification algorithm, as the classification of SUBSCRIBED and UNSUBSCRIBED is categorical.The K-Nearest Neighbours algorithm helps in the creation of a model that classifies observations through the k nearest points to the observation of interest. As the classification of an observation is dependent on its distance to observations of the training set, it is important to standardize the data. With the recipe, we can specify that the predictors are the number of total hours played and the age of the player. However, it is important to note that while the numeric value for the age of a player is often considerably less than the total number of hours played. Furthermore, through Figure 1, we saw that the majority of observations for total number of hours played is in the 0-20 hour bins, whereas Figure 2 shows us that the center of age is easily in the 25+ range. Thus, when creating the recipe, we scale and center our data.

However, it is difficult to evaluate the performance of our prediction model as we do not have the opportunity to collect new observations for testing. Therefore, we will split the original dataset to a training set, for the creation of the model, and a testing set, to evaluate its performance. With this distance-based algorithm, we can plot the number of total hours played and age of all the data in our training set and compute the distance between each point to the total hours played and age of the new observation, choosing the k-nearest points. In this report, we have decided to do a 75/25 split for the training set and testing set respectively, as this is a common division used. To keep this random shuffling and splitting consistent, we will use set.seed(). Additionally, we use cross-validation to evaluate the model's performance across multiple subsets of the dataset. K-fold cross-validation divides a dataset into subsets, allowing each subset to be used as both training and testing data at different stages. This method helps reduce overfitting and ensures our model generalizes well to unseen data.

```r
# add library
library(tidymodels)

set.seed(4235)
players_split <- initial_split(wrangled_players_data, prop = 0.75, strata = subscribe)
players_training <- training(players_split)
players_testing <- testing(players_split)
```

```r

players_recipe <- recipe(subscribe ~ played_hours + Age,
                        data = players_training) |>
  step_scale(all_predictors()) |>
  step_center(all_predictors())
```

Before we can continue to create our model, the KNN Classification algorithm requires one to provide the value for k, such that the model knows the number of nearest points to consider when making a classification. We can use cross-validation to find the accuracy for each k specified, in order to make the most accurate selection. To do this, we can create a specification for our model with the neighbors parameter set to tune().

```r
knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
            set_engine("kknn") |>
            set_mode("classification")
```

In this report, we decided to do a 5-fold cross-validation, resulting in our training data to be split into 5 folds. Furthermore, we will observe the accuracy for k values in the range [1,30], with the subsequent k value incrementing by 2. By specifying this and combining it into a workflow with the specification and recipe defined earlier, the program will automatically split our training data into 5 folds and evaluate the accuracy of each value of k. To choose the most appropriate value, we can plot the accuracy against the number of neighbours and choose the value for k with the highest accuracy.

```r
set.seed(1234)
players_vfold <- vfold_cv(players_training, v = 5, strata = subscribe)

k_vals <- tibble(neighbors = seq(from = 1, to = 40, by = 2))

knn_results <- workflow() |> add_recipe(players_recipe) |> add_model(knn_spec) |> tune_grid(resamples = players_vfold, grid = k_vals) |> collect_metrics()

accuracies <- knn_results |>
                 filter(.metric == 'accuracy')

cross_val_plot <- ggplot(accuracies, aes(x = neighbors, y = mean)) + 
  geom_point(color = "red") + 
  geom_line(color = "blue") +
  labs(
    title = "Figure 3:Accuracy vs Number of Neighbors (k)",
    x = "Number of Neighbors (k)",
    y = "Accuracy Estimate") +
theme(text = element_text(size = 20))

cross_val_plot
```

With this, we see that k=30 is one of the values that provide the highest accuracy. With the best k-value determined, we can now finally create our prediction model by fitting our specification and recipe onto our training data.

```r
knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = 11) |>
  set_engine("kknn") |>
  set_mode("classification")

knn_fit <- workflow() |>
  add_recipe(players_recipe) |>
  add_model(knn_spec) |>
  fit(data = players_training)
```

With our prediction model created, we can now evaluate its performance by making predictions with the testing data to evaluate whether our model correctly classifies the subscription status of the player.

```r
players_test_predictions <- predict(knn_fit, players_testing) |>
  bind_cols(players_testing)

confusion <- players_test_predictions |>
             conf_mat(truth = subscribe, estimate = .pred_class)

confusion

players_test_predictions |>
  metrics(truth = subscribe, estimate = .pred_class) |>
  filter(.metric == "accuracy")

players_test_predictions |>
  precision(truth = subscribe, estimate = .pred_class, event_level = "first")

players_test_predictions |>
  recall(truth = subscribe, estimate = .pred_class, event_level = "first")
```

Based on the confusion matrix presented above, 25 cases were correctly identified as TRUE, 11 cases were mistakenly identified as TRUE, 1 case was correctly identified as FALSE, and 12 cases were mistakenly identified as FALSE. This suggests that the model identifies TRUE cases relatively well (high true positives value), and there is some room for improvement in correctly identifying FALSE cases, as seen in the false positive and false negative counts.



The analysis reveals that a player’s age and total gameplay hours both contribute meaningfully to predicting their subscription status for a gaming newsletter. The K-Nearest Neighbors (KNN) algorithm was used for classification, with cross-validation identifying as the optimal number of neighbors. The model performed well in identifying subscribed players (TRUE cases) with 25 correct classifications, but struggled with unsubscribed players (FALSE cases), showing 11 false positives and 12 false negatives.

The findings align with our assumptions that dedicated players (those with high gameplay hours) and younger players are more likely to subscribe to gaming-related content. It’s not surprising that total hours played was a strong predictor, given its direct connection to player engagement. However, the challenges in correctly classifying unsubscribed players highlight the nuances of human behavior, newsletter subscription may also depend on factors beyond age and hours played, like individual preferences or external influences. For example, A highly engaged player may still choose not to subscribe simply because they dislike newsletters or prefer getting information elsewhere. Or A player who has barely played may still subscribe due to external influence, such as a friend recommending it or an appealing promotional offer.

These insights could help game developers and marketers tailor their outreach strategies. They could focus promotional efforts on younger, high-engagement players to reduce false positives, ensuring their efforts are better targeted and more effective. Predicting subscription status could also help companies optimize newsletter content and improve customer retention.
